отображение и численные результаты:

```python
f1 = 1
f2 = 50
fs = 1000
dt = 1/(0.2*2*f2) #min
t = np.linspace(0, 1, fs)
Z = np.sin(2*np.pi*f1*t) + 0.2*np.sin(2*np.pi*f2*t)

plt.figure(figsize=(10, 6))
plt.plot(t, Z)
plt.title("синал")
plt.xlabel("время (с)")
plt.ylabel("амплитуда")
plt.grid()
plt.show()
```

```python
fft_values = np.fft.fft(Z)
frequencies = np.fft.fftfreq(len(Z), 1/fs)
psd = (np.abs(fft_values) ** 2) / len(Z)

# 3. Ограничение частот до 60 Гц
mask = (frequencies >= 0) & (frequencies <= 60)  # Только положительные частоты до 60 Гц
spectral_power = np.sum(psd[mask]) * (frequencies[1] - frequencies[0])
# 4. Построение графика
plt.figure(figsize=(10, 6))
plt.plot(frequencies[mask], psd[mask] / spectral_power, 'ro', label="Численные результаты")
plt.plot([1, 50], [0.961, 0.038], '.', label="Аналитические результаты") 
plt.text(1, 0.961, ' (1 Гц, 0.961)', fontsize=12, color='red', verticalalignment='bottom')
plt.text(50, 0.038, ' (50 Гц, 0.038)', fontsize=12, color='red', verticalalignment='bottom')
plt.title("cпектральная плотность мощности (до 60 Гц)")
plt.xlabel("частота (Гц)")
plt.ylabel("амплитуда")
plt.grid()
plt.legend()
plt.show()

print(f"Спектральная мощность в диапазоне до 60 Гц: {spectral_power:.2f}")
```

```python
Czz_num= np.zeros((1000))
def ZZ(tt):
    # fs = 1000
    # t = np.linspace(0, 1, fs)
    return np.sin(2*np.pi*f1*tt) + 0.2*np.sin(2*np.pi*f2*tt)

dt = t[2]-t[1]
for i in range(len(Z)):
    Czz_num[i] = np.sum(ZZ(t)*ZZ(t+t[i]))*dt
```

```python
Czz = 1/2*np.cos(2*np.pi*f1*t)+0.2**2/2*np.cos(2*np.pi*f2*t)
```

```python
plt.plot(t, Czz_num, label="Численная автокорреляция (Czz_num)", color='blue', linestyle='-', linewidth=3, alpha=1)
plt.plot(t, Czz, label="Аналитическая автокорреляция (Czz)", color='red', linestyle='-', linewidth=1, alpha=1)

plt.title("Сравнение численной и аналитической автокорреляции", fontsize=16)
plt.xlabel("Время (t)", fontsize=14)
plt.ylabel("Автокорреляция (Czz)", fontsize=14)

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12, loc='upper right')

plt.tight_layout()
plt.show()
```
