```python
import numpy as np
import matplotlib.pyplot as plt
import numba as nb
import sympy as sp
```

Проверка свойств оконных преобразований

```python
N = 10000
x = np.linspace(-100, 100, N)
zero_znach = 1
protya = 3

windows = np.heaviside(x + protya/2, zero_znach) - np.heaviside(x  - protya/2, zero_znach)
```

```python
plt.plot(x, windows)
plt.xlim(-5, 5)
plt.show()
```

```python
dx = x[1] - x[0]
k = np.fft.fftshift(np.fft.fftfreq(N, d=dx)) 
fft_windows = np.fft.fftshift(np.fft.fft(windows))
plt.plot(k, np.abs(fft_windows)**2)
plt.xlim(-2, 2)
plt.show()
```

```python
signal_test_1 = np.sin(2*np.pi*x)
plt.plot(x, signal_test_1)
plt.xlim(-5, 5)
plt.show()
```

```python
fft_signal_test_1= np.fft.fftshift(np.fft.fft(signal_test_1))
plt.plot(k, np.abs(fft_signal_test_1)**2)
plt.xlim(-2, 2)
plt.show()
```

```python
signal_test_2 = signal_test_1*windows
plt.plot(x, signal_test_2)
plt.xlim(-5, 5)
plt.show()
```

```python
fft_signal_test_2 = np.fft.fftshift(np.fft.fft(signal_test_2))
plt.plot(k, np.abs(fft_signal_test_2)**2)
plt.xlim(-2, 2)
plt.show()
```

```python
W = 3/2
gaus_win = (1/(np.sqrt(W*np.sqrt(2*np.pi))))*np.exp(-(x**2)/(4*W**2))
plt.plot(x, gaus_win)
plt.xlim(-5, 5)
plt.show()
```

```python
signal_test_3 = signal_test_1*gaus_win
plt.plot(x, signal_test_3)
plt.xlim(-5, 5)
plt.show()
```

```python
fft_signal_test_3 = np.fft.fftshift(np.fft.fft(signal_test_3))
plt.plot(k, np.abs(fft_signal_test_3)**2)
plt.xlim(-2, 2)
plt.show()
```
